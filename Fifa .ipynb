```python
#Number of players from countries 
data = list(fifa['nationality'].value_counts()[0:5])
characters = list(fifa['nationality'].value_counts()[0:5].keys())

plt.figure(figsize=(5, 5))
plt.bar(characters, data, color=['green', 'pink'])
plt.xticks(rotation=90)
plt.xlabel('Numbers of Countries')
plt.ylabel('Number of Players')
plt.title('Playersfrom countries')
plt.show()
```
